```python
# data_analytics.csv

# update the column names to match the names you have in your csv file for data_analytics

# create a pandas dataframe
dfa = pd.read_csv(
    '/content/data_analytics.csv',
    sep=',',
    names = ['Page path and screen class', 'Views', 'Active users', 'View per active user', 'Average engagement time per active user', 'Event count', 'Key events', 'Total revenue'],
    skiprows=1
)

# Display the first few rows of the DataFrame
print(dfa.head())
```

```python
# data_survey.csv

# update the column names to match the names you have in your csv file for data_survey

# create a pandas dataframe
dfs = pd.read_csv(
    '/content/data_survey.csv',
    sep=',',
    names = ['Timstamp', 'Where did you spend your break?', 'On a scale of 1 to 10, how satisfied are you overall experience?'],
    skiprows=1
)

# Display the first few rows of the DataFrame
print(dfs.head())
```

```python
#  how many rows and columns are in the dataframe dfa?
# how many rows and columns are in the dataframe dfs?
```

```python
# Calculate and print dimensions for dfa
dfa_rows, dfa_cols = dfa.shape
print(f"The dataframe 'dfa' has {dfa_rows} rows and {dfa_cols} columns.")

# Calculate and print dimensions for dfs
dfs_rows, dfs_cols = dfs.shape
print(f"The dataframe 'dfs' has {dfs_rows} rows and {dfs_cols} columns.")
```

```python
# Analytics

# Do one or more explorations with dfa
# no interaction is required
dfa_clean = dfa.dropna(subset=['Views']).copy() # drops row where "Views" is NaN

#Filter out rows where 'Views' is actually the column header text or metadata
dfa_clean= dfa_clean[pd.to_numeric(dfa_clean['Views'], errors='coerce').notnull()]


dfa_clean['Views'] = pd.to_numeric(dfa_clean['Views'])  # converts "Views" to numeric just in case it's stored as a string

# make a blank canvas
plt.figure(figsize=(10,6))
plt.bar(dfa_clean['Page path and screen class'], dfa_clean['Views'], color = 'blue')

# make a title starting with 'Figure 1:' Fill in a relevant title
plt.title('Figure 1: Page Views by Screen Class')
plt.xlabel('Screen')
plt.ylabel('Number of Views')

#Rotate the x-axis labels so they are readable
plt.xticks(rotation=45, ha ='right')

plt.tight_layout()

plt.show()

```

```python
# Survey

# Do one or more explorations with dfs
# no interaction is required

# make a blank canvas
plt.figure(figsize=(10,6))

plt.bar(dfs['Where did you spend your break?'],
        dfs['On a scale of 1 to 10, how satisfied are you overall experience?'],
        color='green')

# make a title starting with 'Figure 2:' Fill in a relevant title
plt.title('Figure 2:Break Satisfaction Scores by Location')
plt.xlabel('Location')
plt.ylabel('Satisfaction Score (1-10)')
plt.ylim(0,11)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


```
